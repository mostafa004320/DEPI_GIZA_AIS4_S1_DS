# Write a Python program to remove an item from a given set

## Example
my_set: {1, 2, 3, 4, 5}

Output: {1, 2, 4, 5}

```python
# Remove an item from a set
my_set = {1, 2, 3, 4, 5}
item_to_remove = 3

# Use discard to avoid KeyError if the item isn't found
my_set.discard(item_to_remove)

print(my_set)  # {1, 2, 4, 5}
```

# Write a Python program to removes multiple items from a given set

## Example
my_set: {1, 2, 3, 4, 5}

items to be removed: {1, 4}

Output: {2, 3, 5}

```python
# Remove multiple items from a set
my_set = {1, 2, 3, 4, 5}
items_to_remove = {1, 4}

# Option 1: difference_update (in-place)
my_set.difference_update(items_to_remove)

print(my_set)  # {2, 3, 5}
```

# Write a Python program to find the index of an item in a tuple.

## Example
my_tuple: ('apple', 'banana', 'orange', 'grape', 'banana')

index of 'orange' is: 2

```python
# Find the index of an item in a tuple
my_tuple = ('apple', 'banana', 'orange', 'grape', 'banana')
item = 'orange'

# tuple has index() method (returns first occurrence)
idx = my_tuple.index(item)

print(f"index of '{item}' is: {idx}")
```

# Create a function that Calculates the Sum of Even Numbers in a List
## Example
numbers: [1, 2, 3, 4, 5, 6]

Output: 12

```python
# Write your answer here
"""
This function takes a list of numbers and returns the sum of all even numbers in the list
Example:
Input: [1, 2, 3, 4, 5, 6]
Output: 12
"""
def sum_even_numbers(numbers):
    total = 0
    for i in numbers:
        if i % 2 == 0:
            total += i
    return total
numbers = [1, 2, 3, 4, 5, 6]
print(sum_even_numbers(numbers))
```

# Create a function that takes a list as input and return reversed list

## Example
lst: [1, 2, 3]

Output: [3, 2, 1]

```python
# Write your answer here
"""
This function takes a list as input and returns the reversed version of the list
Example:
Input: [1, 2, 3,4]
Output: [4,3, 2, 1]
"""
def reverse_list(lst):
    return lst[::-1]
lst = [1, 2, 3]
print(reverse_list(lst))
```

# create a function that calculates the median of a list of numbers (a median is the middle element in sorted list of numbers)

## Example
numbers: [5, 2, 7, 3, 1]

Output: 3

```python
# Write your answer here
"""
This function takes a list of numbers, sorts it,
and returns the median (the middle element of the sorted list)
If the list length is even, it returns the average of the two middle numbers

Example:
Input: [5, 2, 7, 3, 1]
Output: 3
"""
def median(numbers):
    numbers = sorted(numbers)
    n = len(numbers)
    mid = n // 2
    if n % 2 != 0:
        return numbers[mid]
    else:
        return (numbers[mid - 1] + numbers[mid]) / 2
numbers = [5, 2, 7, 3, 1]
print(median(numbers))
```
